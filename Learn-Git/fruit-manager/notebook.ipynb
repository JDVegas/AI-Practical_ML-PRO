```python
# Create inventory for our growing (exploitation)
inventaire = {
    "bananes": 120
    , "mangues": 85
    , "ananas": 45
    , "noix de coco": 60
    , "papayes": 30
}
```

```python
# Function that display the inventory 
def afficher_inventaire(inventaire):
    print("Inventaire actuel de la plantation :")
    for fruit, quantite in inventaire.items():
        print(f"- {fruit.capitalize()} : {quantite} unité")
```

```python
# Demo
afficher_inventaire(inventaire)
```

```python
# Function to harvest a spceific amount of a fruit
def recolter(inventaire, fruit, quantite):
    inventaire[fruit] = inventaire.get(fruit, 0) + quantite
    print(f"\nRécolté {quantite} {fruit} supplémentaires !")
```

```python
# Demo
recolter(inventaire, "bananes", 10)

afficher_inventaire(inventaire)
```
